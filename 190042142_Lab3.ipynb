```python
# Task1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Real_estate.csv")
dataplot = sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

# New Section

```python
# Task2
X = np.array(data["Distance station"])
Y = np.array(data["Longitude"])

plt.scatter(X,Y,c= "blue")
plt.show()
```

```python
# Task3
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=42)

X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)

reg = LinearRegression()
reg.fit(X_train,y_train)
reg.predict(X_test)


print(reg.coef_)
print(reg.intercept_)
print(reg.score(X_test,y_test))


```

```python
# Task 4
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=42)

X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)

reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)


print(mean_absolute_error(y_test,y_pred))

print(mean_squared_error(y_test,y_pred))

print(r2_score(y_test,y_pred))
```

```python
# Task 5
plt.scatter(X_test,y_test, c ="blue",marker='x')
plt.plot(X_test,y_pred,c="red")
plt.vlines(X_test,y_test,y_pred,alpha =0.3)
plt.show()
```

```python
# Task 6

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
sumx = sumy = sumxy = sumx2 = 0
n = len(X_train)

for i in range(n):
  sumx += X_train[i]
  sumy += y_train[i]
  sumxy += X_train[i] * y_train[i]
  sumx2 += X_train[i] ** 2

slope = (n*sumxy - sumx*sumy)/(n*sumx2 - (sumx)**2)
intercept = (sumy/n) - (slope * (sumx/n))


y_pred3 = intercept + (slope * X_test)
print(mean_absolute_error(y_test, y_pred3))
print(mean_squared_error(y_test, y_pred3))
print(r2_score(y_test, y_pred3))

plt.scatter(X_test, y_test, c='blue', marker='x')
plt.plot(X_test, y_pred3, c='red')
plt.show()
```

```python
# Task 8

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
p = np.polyfit(X_train, y_train, 1)
print(p[0])
print(p[1])

y_pred2 = np.polyval(p, X_test)

print(mean_absolute_error(y_test, y_pred2))
print(mean_squared_error(y_test, y_pred2))
print(r2_score(y_test, y_pred2))

plt.scatter(X_test, y_test, c='blue', marker='x')
plt.plot(X_test, y_pred2, c='red')
plt.show()
```
